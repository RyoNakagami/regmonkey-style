```python
import plotly.express as px
from IPython.display import Image, display
df = px.data.tips()
fig = px.box(df, x="time", y="total_bill")

# output path
output_path_iris = "./output/regmonkey_boxplot_plotly_2groups.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```

```python
import matplotlib.pyplot as plt
import numpy  as np


# Create a box plot using Matplotlib
plt.figure(figsize=(10, 6))

# Group data by 'day' and 'smoker'
grouped = df.groupby(['time'])['total_bill'].apply(lambda x: list(x))


# Create box plot
fig, ax = plt.subplots(figsize=(5, 6))
ax.boxplot(
    grouped,
    widths=0.5,
    tick_labels=grouped.index 
)

# Set plot attributes
ax.set_title("RegStyle Box Plot")
ax.set_xlabel("Category")
ax.set_ylabel("Values")
plt.suptitle("")  # Remove the automatic 'Boxplot grouped by ...' title

# Show plot
plt.show()
```

```python
fig = px.box(df, x="day", y="total_bill")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default

# output path
output_path_iris = "./output/regmonkey_boxplot_plotly_4groups.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```

```python
fig = px.box(df, x="day", y="total_bill", color="smoker")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default

# output path
output_path_iris = "./output/regmonkey_boxplot_plotly_4groups_with_color.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```

```python
import plotly.graph_objects as go

x = ['day 1', 'day 1', 'day 1', 'day 1', 'day 1', 'day 1',
     'day 2', 'day 2', 'day 2', 'day 2', 'day 2', 'day 2']

fig = go.Figure()

fig.add_trace(go.Box(
    y=[0.2, 0.2, 0.6, 1.0, 0.5, 0.4, 0.2, 0.7, 0.9, 0.1, 0.5, 0.3],
    x=x,
    name='kale',
))
fig.add_trace(go.Box(
    y=[0.6, 0.7, 0.3, 0.6, 0.0, 0.5, 0.7, 0.9, 0.5, 0.8, 0.7, 0.2],
    x=x,
    name='radishes',
))
fig.add_trace(go.Box(
    y=[0.1, 0.3, 0.1, 0.9, 0.6, 0.6, 0.9, 1.0, 0.3, 0.6, 0.8, 0.5],
    x=x,
    name='carrots',
))

fig.update_layout(
    yaxis_title='normalized moisture',
    boxmode='group' # group together boxes of the different traces for each value of x
)
# output path
output_path_iris = "./output/regmonkey_boxplot_plotly_3groups_with_color.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```
