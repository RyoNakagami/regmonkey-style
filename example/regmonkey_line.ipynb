```python
import plotly.express as px
from IPython.display import Image, display

df = px.data.gapminder()

# Filter the DataFrame
nice_countries = df['country'].unique()[:9]
filtered_df = df.loc[df['country'].isin(nice_countries)]

fig = px.line(
    filtered_df,
    x="year",
    y="lifeExp",
    title="Life expectancy in Canada",
    color='country',
    width=800,
    height=600,
)

# output path
output_path_iris = "./output/regmonkey_line_plotly_default_life_expectancy.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```

```python
import plotly.graph_objects as go


# Create a line plot using Plotly Graph Objects
fig = go.Figure()

# Add traces for each country
for country in filtered_df['country'].unique():
    country_data = filtered_df[filtered_df['country'] == country]
    fig.add_trace(go.Scatter(
        x=country_data['year'],
        y=country_data['lifeExp'],
        mode='lines',
        name=country
    ))

# Update layout
fig.update_layout(
    title="Life expectancy in Canada",
    xaxis_title="Year",
    yaxis_title="Life Expectancy",
    width=800,
    height=600
)

# Output path
output_path_iris = "./output/regmonkey_line_plotly_life_expectancy_go_scatter.png"
fig.write_image(output_path_iris, engine="kaleido")
display(Image(filename=output_path_iris))
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Create a line plot using Matplotlib
fig, ax = plt.subplots(figsize=(8, 6))

# Plot each country separately
for country, group in filtered_df.groupby('country'):
    ax.plot(group['year'], group['lifeExp'], label=country)

# Set the title and labels
ax.set_title("Life expectancy in Canada")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")

# Add a legend
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0.)

# Save the plot as an image
output_path_iris = "./output/regmonkey_line_matplotlib_life_expectancy.png"
plt.savefig(output_path_iris, bbox_inches='tight')

# Display the plot
plt.show()
```
